#### Item a)

```python
item_a_1_data = pd.read_table('item_a-1.txt', sep=' ', header=None)
item_a_2_data = pd.read_table('item_a-2.txt', sep=' ', header=None)
item_a_3_data = pd.read_table('item_a-3.txt', sep=' ', header=None)

plt.figure(figsize=(10, 6))
plt.plot(item_a_1_data[0], item_a_1_data[1], label=r'$\dot{x}(0)=0.10$')
plt.plot(item_a_2_data[0], item_a_2_data[1], label=r'$\dot{x}(0)=0.25$')
plt.plot(item_a_3_data[0], item_a_3_data[1], label=r'$\dot{x}(0)=0.50$')
plt.xlabel(r'$x(t)$')
plt.ylabel(r'$\dot{x}(t)$')
plt.title('Espaço de fase (Item a)')
plt.legend()
plt.show()
```
